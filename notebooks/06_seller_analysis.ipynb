# Seller Analysis

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

pd.set_option("display.max_columns", None)
pd.set_option("display.max_rows", None)


sns.set_style("whitegrid")
plt.rcParams["figure.figsize"] = (10, 6)
plt.rcParams["font.size"] = 12
```

```python
ecom_data_df = pd.read_parquet("../data/olist_ecom.parquet")
```

## Aggregate Seller Performance Metrics

```python
seller_summary = (
    ecom_data_df.groupby("seller_id")
    .agg(
        total_sales=("price", "sum"),
        total_freight=("freight_value", "sum"),
        total_orders=("order_id", "nunique"),
        avg_delivery_time=(
            "total_delivery_time",
            "mean",
        ),
        avg_delivery_vs_estimate=(
            "delivery_vs_estimate",
            "mean",
        ),  # if value is negative thats mean is good
    )
    .reset_index()
)
```

```python
# calculate total revenue(price(total_sales) + freight(total_freight))
seller_summary["total_revenue"] = (
    seller_summary["total_sales"] + seller_summary["total_freight"]
)
```

```python
print(f"Shape of seller_summary: {seller_summary.shape}")
display(seller_summary.sample(5))
```

## Identify Top/Bottom Sellers (Table Output)

```python
top_sellers_by_revenue = seller_summary.sort_values(
    by="total_revenue", ascending=False
).head(10)
```

```python
print("Top 10 seller based total revenue")
display(top_sellers_by_revenue[["seller_id", "total_revenue", "total_orders"]])
```

```python
top_sellers_by_orders = seller_summary.sort_values(
    by="total_orders", ascending=False
).head(10)
```

```python
print("Top 10 seller based total orders")
display(top_sellers_by_orders[["seller_id", "total_revenue", "total_orders"]])
```

## Visualize Top Sellers by Total Revenue

```python
plt.figure(figsize=(12, 7))
sns.barplot(
    x="total_revenue",
    y="seller_id",
    data=top_sellers_by_revenue.sort_values(by="total_revenue", ascending=True),
)
plt.title("Top 10 Sellers by Total Revenue", fontsize=16)
plt.xlabel("Total Revenue (R$)", fontsize=12)
plt.ylabel("Seller ID", fontsize=12)
plt.grid(axis="x", linestyle="--", alpha=0.7)
plt.tight_layout()
plt.savefig("top_10_sellers_revenue.png")
plt.show()
```

## Visualize Top Sellers by Total Orders

```python
plt.figure(figsize=(12, 7))
sns.barplot(
    x="total_orders",
    y="seller_id",
    data=top_sellers_by_orders.sort_values(by="total_orders", ascending=True),
)
plt.title("Top 10 Sellers by Total Orders", fontsize=16)
plt.xlabel("Total Orders (R$)", fontsize=12)
plt.ylabel("Seller ID", fontsize=12)
plt.grid(axis="x", linestyle="--", alpha=0.7)
plt.tight_layout()
plt.savefig("top_10_sellers_orders.png")
plt.show()
```

## Explore Correlations (Optional/Advanced Visualizations)

```python
# Scatter plot: Total Orders vs. Average Delivery Time
plt.figure(figsize=(10, 6))
sns.scatterplot(
    x="total_orders", y="avg_delivery_time", data=seller_summary, alpha=0.6, s=100
)
plt.title("Total Orders vs. Average Delivery Time per Seller", fontsize=16)
plt.xlabel("Total Orders", fontsize=12)
plt.ylabel("Average Delivery Time (Days)", fontsize=12)
plt.grid(True, linestyle="--", alpha=0.6)
plt.tight_layout()
plt.savefig("orders_vs_delivery_time.png")
plt.show()
```

```python
# Scatter plot: Total Orders vs. Average Delivery Performance vs. Estimate
plt.figure(figsize=(10, 6))
sns.scatterplot(
    x="total_orders",
    y="avg_delivery_vs_estimate",
    data=seller_summary,
    alpha=0.6,
    s=100,
)
plt.title(
    "Total Orders vs. Average Delivery Performance vs. Estimate per Seller", fontsize=16
)
plt.xlabel("Total Orders", fontsize=12)
plt.ylabel("Avg. Delivery vs. Estimate (Days, negative = early)", fontsize=12)
plt.axhline(0, color="red", linestyle="--", linewidth=0.8)
plt.grid(True, linestyle="--", alpha=0.6)
plt.tight_layout()
plt.savefig("orders_vs_delivery_vs_estimate.png")
plt.show()
```
